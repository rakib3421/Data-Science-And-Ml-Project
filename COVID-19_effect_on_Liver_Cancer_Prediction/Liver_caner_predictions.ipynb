```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('file1.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.drop(['Type_of_incidental_finding','Surveillance_effectiveness','Mode_of_surveillance_detection','Date_incident_surveillance_scan','Time_diagnosis_1st_Tx','Time_MDM_1st_treatment','Time_decisiontotreat_1st_treatment','Months_from_last_surveillance','Unnamed: 0'],axis=1,inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['Treatment_grps'] = df['Treatment_grps'].fillna('Supportive care')
```

```python
df['Surveillance_programme'] = df['Surveillance_programme'].fillna('U')
```

```python
df['PS'] = df['PS'].fillna(df['PS'].median())
```

```python
df['Prev_known_cirrhosis'] = df['Prev_known_cirrhosis'].fillna('Y')
```

```python
df.isnull().sum()
```

```python
x = df.drop(['Cancer'],axis=1)
y = df['Cancer']
```

```python
x
```

```python
x.columns
```

```python
list1 = ['Bleed', 'Mode_Presentation', 'Gender', 'Etiology', 'Cirrhosis','HCC_TNM_Stage','HCC_BCLC_Stage','Treatment_grps','Alive_Dead', 'Surveillance_programme','Prev_known_cirrhosis']
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for i in list1:
    x[i] = le.fit_transform(x[i])
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
```

```python
gnb.fit(x_train,y_train)
```

```python
gnb.score(x_test,y_test)*100
```

```python
pred = gnb.predict(x_test)
```

```python
pred
```

```python
y_test
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,pred))
print(confusion_matrix(y_test,pred))
```

```python
y_test.value_counts()
```

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
```

```python
logmodel.fit(x_train,y_train)
```

```python
logmodel.score(x_test,y_test)
```

```python
pred = logmodel.predict(x_test)
```

```python
pred
```
